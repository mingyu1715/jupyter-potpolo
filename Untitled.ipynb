```python
import csv
import matplotlib.pyplot as plt

file_names = [
    "highschool2022.csv",
    "highschool2021.csv",
    "highschool2020.csv",
    "highschool2019.csv",
    "highschool2018.csv",
    "highschool2017.csv",
    "highschool2016.csv",
    "highschool2015.csv",
    "highschool2014.csv",
    "highschool2013.csv",
    "highschool2012.csv",
    "highschool2011.csv",
    "highschool2010.csv",
    "highschool2009.csv",
    "highschool2008.csv",
]

data0 = []
for file_name in file_names:
    with open(file_name, encoding="ANSI") as file:
        reader = csv.reader(file, delimiter=",")
        next(reader)
        next(reader)
        next(reader)
        data0.append(list(reader))

list1 = [
    "2008",
    "2009",
    "2010",
    "2011",
    "2012",
    "2013",
    "2014",
    "2015",
    "2016",
    "2017",
    "2018",
    "2019",
    "2020",
    "2021",
    "2022",
]

a1 = [0] * 15
a2 = [0] * 15
a3 = [0] * 15

def get_total_students_high(data,col_idx):
    total_students = 0
    for row in data:
        if row[4] in '고등학교':
            total_students += int(row[col_idx])
        return total_students

for i in range(15):
    if i == 0:
        index_a1 = 24 
    elif i == 1:
        index_a1 = 23
    else:
        index_a1 = 28

    a1[i] = get_total_students_high(data0[i], index_a1)

def get_total_students_middle(data,col_idx):
    total_students = 0
    for row in data:
        if row[4] in '중학교':
            total_students += int(row[col_idx])
        return total_students

for i in range(15):
    if i == 0:
        index_a1 = 24 
    elif i == 1:
        index_a1 = 23
    else:
        index_a1 = 28

    a2[i] = get_total_students_middle(data0[i], index_a1)

def get_total_students_elementary(data,col_idx):
    total_students = 0
    for row in data:
        if row[4] in '초등학교':
            total_students += int(row[col_idx])
        return total_students

for i in range(15):
    if i == 0:
        index_a1 = 24 
    elif i == 1:
        index_a1 = 23
    else:
        index_a1 = 28

    a3[i] = get_total_students_elementary(data0[i], index_a1)

a1.reverse()
a2.reverse()
a3.reverse()

plt.rc("font", family="Malgun Gothic")
plt.title("연도별 반평균 고등학교 학생수")

plt.plot(list1, a1, marker="o", label="고등학교")  # 연도와 a1에 대한 꺾은선 그래프 생성
plt.plot(list1, a2, marker="o", label="중학교")
plt.plot(list1, a3, marker="o", label="초등학교")

plt.xticks(range(15), list1, rotation=90)  # 가로축 각도는 90도
plt.xlabel("연도")
plt.ylabel("학생 수")
plt.legend()
plt.show()

print(a1)
print(a2)
print(a3)
```
